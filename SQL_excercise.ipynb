## Aggregates

1. What is the average BACDistance (in Km)? Rename the ouptut column to "avg_bac_distance."

````python
# Write your SQL query in this string
sql = 'SELECT AVG(BACDDistance) as avg_bac_distance FROM school_details'

# Read the sql query into pandas to view the results
df = pd.read_sql_query(sql, engine)
df.head()
````

## GROUP BY

2. What is the average BACDistance (in Km) by School Level? Sort the results from highest average to lowest average.

````python
# Write your SQL query in this stringSchool Level
sql = 'SELECT AVG(BACDDistance) AS avg_bacd_distance FROM School_details GROUP BY SchoolLevel ORDER BY avg_bacd_distance ASC'


# Read the sql query into pandas to view the results
df = pd.read_sql_query(sql, engine)
df.head()
````

## Having

3. Repeat the previous query, but only display results where the average distance is at least 10 km.

````python
# Write your SQL query in this string

sql = 'SELECT AVG(BACDDistance) AS avg_bacd_distance FROM School_details GROUP BY SchoolLevel HAVING AVG(BACDDistance)>10 ORDER BY avg_bacd_distance ASC'

# Read the sql query into pandas to view the results
df = pd.read_sql_query(sql, engine)
df.head(10)
````

## Joins

4. Join the two tables together (school_details and school_details). Start by displaying all of the columns, but limit the resulting rows to just 2.

````python
# Write your SQL query in this string
sql = 'SELECT * FROM school_details as S INNER JOIN school_facility_details as F  ON F.EstablishmentCode = S.EstablishmentCode LIMIT 2'

# Read the sql query into pandas to view the results
df = pd.read_sql_query(sql, engine)
df.head()
````

5. Now, select just the schools where the SchoolLevel is "PS" that do not have libraries.

````python
# Write your SQL query in this string
sql = "SELECT F.EstablishmentName FROM school_details as S LEFT JOIN school_facility_details as F ON S.EstablishmentCode = F.EstablishmentCode WHERE SchoolLevel = 'PS' AND IsLibraryAvailable ='Not Available'"
# Read the sql query into pandas to view the results
df = pd.read_sql_query(sql, engine)
df
````

6. Build off the previous query, but order the results alphabetically by name.

````python
# Write your SQL query in this string 
sql = "SELECT F.EstablishmentName FROM school_details as S LEFT JOIN school_facility_details as F ON S.EstablishmentCode = F.EstablishmentCode WHERE SchoolLevel = 'PS' AND IsLibraryAvailable ='Not Available' ORDER BY F.EstablishmentName ASC"

# Read the sql query into pandas to view the results
df = pd.read_sql_query(sql, engine)
df.head()
````
