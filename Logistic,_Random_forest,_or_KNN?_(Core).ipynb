```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, recall_score, precision_score, \
f1_score, classification_report, confusion_matrix
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.multiclass import OneVsRestClassifier, OneVsOneClassifier
```

```python
df = pd.read_csv('/content/drive/MyDrive/wine.data')
df
```

```python
import pandas as pd
# list the names of the columns from the wine.name file 
col_names = ['Class label', 'Alcohol', 'Malic acid', 'Ash', 'Alcalinity of ash', 'Magnesium', 'Total phenols',
'Flavanoids', 'Nonflavanoid phenols', 'Proanthocyanins','Color intensity', 'Hue', 'OD280/OD315 of diluted wines','Proline']
# define the url and make sure to specify the wine.data file at the end
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data'
# read in the file from the url and include the list of names as your column names
df = pd.read_csv(url, header = None, names = col_names)
df
```

```python
df.info()
```

```python
print('Missing data:\n', df.isna().sum())
print('\n')
print('Duplicates:', df.duplicated().sum())
```

```python
X = df.drop(columns = ['Class label'])
y = df['Class label']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, stratify=y)
```

# LogisticRegression

```python
logreg_params = {
    'penalty': ['l2',],
               'C': [1, 2, 3, 4, 5, 10, 20, 50, 100, 500, 1000],
               'dual': [False],
               'multi_class': ['ovr']
               }
model = LogisticRegression(random_state = 42)
grid = GridSearchCV(model, logreg_params)
```

```python
grid.fit(X_train, y_train)
```

```python
best_lr = grid.best_estimator_

lr_train_preds = best_lr.predict(X_train)
lr_test_preds = best_lr.predict(X_test)
print('Best LR Model Train Metrics')
print(classification_report(y_train, lr_train_preds))
print('\nBest LR Model Test Metrics')
print(classification_report(y_test, lr_test_preds))

print('Classification Report for LR Training Set')
lr_train_report = classification_report(y_train, lr_train_preds)
print(lr_train_report)

print('Confusion Matrix for LR Training Set')
lr_train_conf_mat = confusion_matrix(y_train, lr_train_preds)
print(lr_train_conf_mat)

print('\n\n')

print('Classification Report for LR Testing Set')

lr_test_report = classification_report(y_test, lr_test_preds)
print(lr_test_report)

print('Confusion Matrix for LR Testing Set')
lr_test_conf_mat = confusion_matrix(y_test, lr_test_preds)
print(lr_test_conf_mat)
```

```python
model = LogisticRegression(C = 1, dual = False, multi_class = 'ovr', penalty= 'l2', random_state = 42)
ovo = OneVsOneClassifier(model)
ovo.fit(X_train, y_train)
ovo_y_pred = ovo.predict(X_test)
ovo.score(X_train, y_train)
```

# KNN

```python
scaler = StandardScaler()
knn = KNeighborsClassifier()

knn_pipe = make_pipeline(scaler, knn)

knn_pipe.fit(X_train, y_train)
```

```python
train_preds = knn_pipe.predict(X_train)
test_preds = knn_pipe.predict(X_test)

print('train accuracy:', accuracy_score(y_train, train_preds))
print('\n')
print('test accuracy:', accuracy_score(y_test, test_preds))
```

```python
logreg_params = {
    'penalty': ['l2',],
               'C': [1, 2, 3, 4, 5, 10, 20, 50, 100, 500, 1000],
               'dual': [False],
               'multi_class': ['ovr']
               }
```

```python
params = {'kneighborsclassifier__n_neighbors': range(1,20),
          'kneighborsclassifier__weights': ['uniform','distance'],
          'kneighborsclassifier__p': [2,3,4]}
```

```python
knn_grid = GridSearchCV(knn_pipe, params)

knn_grid.fit(X_train, y_train)
```

```python
knn_grid.best_params_
```

```python
best_knn = knn_grid.best_estimator_
```

```python
train_preds = best_knn.predict(X_train)
test_preds = best_knn.predict(X_test)
print('Best Model Train Metrics')
print(classification_report(y_train, train_preds))
print('\nBest Model Test Metrics')
print(classification_report(y_test, test_preds))
```

```python
print('Classification Report for Training Set')
train_report = classification_report(y_train, train_preds)
print(train_report)

print('Confusion Matrix for Training Set')
train_conf_mat = confusion_matrix(y_train, train_preds)
print(train_conf_mat)

print('\n\n')

print('Classification Report for Testing Set')

test_report = classification_report(y_test, test_preds)
print(test_report)

print('Confusion Matrix for Testing Set')
test_conf_mat = confusion_matrix(y_test, test_preds)
print(test_conf_mat)
```

# RF

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_params = {
              'n_estimators': [10, 100, 300],
              'max_depth': range(2,15),
              'min_samples_split': [2, 3, 4, 5]
              }
rf = RandomForestClassifier(random_state = 42)
rf_grid = GridSearchCV(rf, rf_params)
```

```python
rf_grid.fit(X_train, y_train)
```

```python
rf_grid.best_params_
```

```python
best_rf = rf_grid.best_estimator_
```

```python
train_preds = best_rf.predict(X_train)
test_preds = best_rf.predict(X_test)
print('Best Model Train Metrics')
print(classification_report(y_train, train_preds))
print('\nBest Model Test Metrics')
print(classification_report(y_test, test_preds))
```

```python
print('Classification Report for Training Set')
train_report = classification_report(y_train, train_preds)
print(train_report)

print('Confusion Matrix for Training Set')
train_conf_mat = confusion_matrix(y_train, train_preds)
print(train_conf_mat)

print('\n\n')

print('Classification Report for Testing Set')

test_report = classification_report(y_test, test_preds)
print(test_report)

print('Confusion Matrix for Testing Set')
test_conf_mat = confusion_matrix(y_test, test_preds)
print(test_conf_mat)
```
