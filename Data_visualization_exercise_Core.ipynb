```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
filename = '/content/drive/MyDrive/RailsToTrails_National Count Data_week.xlsx'
df = pd.read_excel('/content/drive/MyDrive/RailsToTrails_National Count Data_week.xlsx')
df
```

```python
df.hist();
```

```python
year_df = df['Year']==2020
print(year_df)
```

```python
plt.barh(df['Year'], df['Bikes (14 counters)']);
plt.xticks(rotation = 45)
plt.xlabel('Year', fontsize = 15)
plt.ylabel('Bikes (14 counters)', fontsize = 15);
plt.ticklabel_format(useOffset=False, style='plain')
plt.title('bikes based on the year', fontsize = 15);
```

```python
plt.barh(df['Year'], df['Counts (31 counters)']);
plt.xticks(rotation = 45)
plt.xlabel('Year', fontsize = 15)
plt.ylabel('Counts (31 counters)', fontsize = 15);
plt.ticklabel_format(useOffset=False, style='plain')
plt.title('bikecounters based on year', fontsize = 15);
```

In this graph i choose the year based on 31 counters


```python
df['Year'].hist();
df['Pedestrians (14 counters)'].hist();
plt.xticks(rotation = 45)
plt.ticklabel_format(useOffset=False, style='plain')
plt.title('Pedestrians based on year', fontsize = 15);
plt.xlabel('Year', fontsize = 15)
plt.ylabel('Pedestrians (14 counters)', fontsize = 15);
```

In this graph i choose the pedestrians and year.

```python
sns.boxplot(x='Year', y='Bikes (14 counters)',data=df,notch=True);
```
