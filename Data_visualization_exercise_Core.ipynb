```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
filename = '/content/drive/MyDrive/RailsToTrails_National Count Data_week.xlsx'
df = pd.read_excel('/content/drive/MyDrive/RailsToTrails_National Count Data_week.xlsx')
df
```

```python
df.hist();
```

```python
df['Year'].hist();
plt.xticks(rotation = 45)
plt.ticklabel_format(useOffset=False, style='plain')

```

```python
df['Counts (31 counters)'].hist();
plt.xticks(rotation = 45)
plt.ticklabel_format(useOffset=False, style='plain')
```

```python
df['Pedestrians (14 counters)'].hist();
plt.xticks(rotation = 45)
plt.ticklabel_format(useOffset=False, style='plain')
```

```python
df['Bikes (14 counters)'].hist();
plt.xticks(rotation = 45)
plt.ticklabel_format(useOffset=False, style='plain')
```

```python
sns.boxplot(x='Year', y='Bikes (14 counters)',data=df,notch=True);
```

```python
df.corr()
```

```python
corr = df.corr()
sns.heatmap(corr, cmap='Blues');
```
